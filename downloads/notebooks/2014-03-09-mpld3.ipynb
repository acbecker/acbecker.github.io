# Making the Connection between LSST and iPython

#### After the adoption of the [Anaconda](https://store.continuum.io/cshop/anaconda/) platform for [LSST](http://www.lsst.org/lsst/) Data [Management's](https://dev.lsstcorp.org/trac) Python distribution, a broad array of packages have now become natively integrated with LSST's Python code.  The [iPython](http://ipython.org/) package in particular opens up a new means of communicating information - plots, figures, and example code - in a way that allows people both internal and external to DM to interact with the results.      An example application is given here, where I have used the [camera geometry](https://dev.lsstcorp.org/trac/ticket/3117) development work of Simon Krughoff and Russ Owen, along with the [matplotlib-D3](https://github.com/jakevdp/mpld3) integration work of Jake Vanderplas, to visualize (and interact with) the LSST focal plane.  While the in-browser interaction capabilities are mostly limited to zoom and pan, there are other developments on the horizon such as [linked brushing](https://github.com/jakevdp/mpld3/blob/master/examples/linked_brush.py) in mpld3, and multiple packages under development to integrate Python and [D3](http://d3js.org/), that  may make this a fundamental means of communication within the project in the near future.

The commands to integrate matplotlib and D3 are as follows:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpld3 import enable_notebook
enable_notebook()
```

And the commands to plot up the LSST focal plane are below:

```python
import lsst.obs.lsstSim.utils as lsstSimUtils
import lsst.afw.cameraGeom.utils as cameraGeomUtils
import eups
lsstSimDir = eups.productDir('obs_lsstSim')
camera = lsstSimUtils.loadCamera(lsstSimDir)
cameraGeomUtils.plotFocalPlane(camera, 3.5, 3.5)
```

Easy!  To interact with the figure, choose the left-right-up-down arrow in the lower left corner of the figure, and this will interpret your mouse gestures as pan/zoom commands.

In this figure depicting the LSST focal plane, the blue squares represent the imaging CCD devices, the red square wavefront sensors, and the green squares the guide sensors.  Dots represent the projection of an evenly-spaced grid at the pupil onto the focal plane; a slight pincushion distortion is visible.  When zooming in, slight offsets can be seen in the positions of the devices.  These are derived from a realistic CCD placement model of Andy Rasumussen.
