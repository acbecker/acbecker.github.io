# Bayesian Data Analysis : Intro

## Terminology and Notation:

$p(\theta | y)$ : Probability of model parameters $\theta$ given data $y$.  Probability of model parameters "conditional" on the data.  Also known as the _posterior_ distribution.

$p(y|\theta)$ : Probabilty of data, given the model with parameters $\theta$.  Also known as the _sampling_ or _data_ distribution.

$p(\theta)$ : _Prior_ distribution of the model parameters.  Typically specified by the modeler (i.e. a degree of freedom in your analysis)

Bayes Rule : $p(\theta|y) \propto  p(\theta) \cdot p(y|\theta)$
 $$ $$
Or : $ln(posterior) \propto  ln(prior) + ln(data\ likelihood)$

Frequentists work with $p(y|\theta)$.  That is, they typically specify a model and look at the $\chi^2$ value of the data.  Different models may be compared based on their $\chi^2$ per degree of freedom, where the degrees of freedom are the number of data points fitted, minus the number of terms fitted.  As you make a model more and more complex, the $\chi^2$ will _always_ go down.  But as you start to **overfit** the problem, the $\Delta \chi^2$ will become smaller and smaller, indicating less and less relevance for the new model parameters.  The predictive power of this model will also decrease, indicating it becomes a high _variance_ estimator.  You always want to trade off **bias** and **variance** when fitting models to data.

For a normal distribution the probability density function is: $f(x, \mu, \sigma) = \frac{1}{\sigma \sqrt{2 * \pi}} e^\frac{(x - \mu)^2}{2 \sigma^2}$.  Looking at e.g. fitting data to a line (y = m x + b), we have $\mu$ determined by the model ($\mu = mx + b$) and $\sigma$ the uncertainties on your data.  Looking at this in a Bayesian context, we have $ln\ p(y|x,n,m,b,\sigma) = -0.5 [ \sum_n$$ \frac{(y_n - (m * x_n + b))^2}{\sigma_n^2} + ln(2\pi\sigma_n^2) ]$.  Since $\sigma_n$ are typically fixed, you can approximate $ln(data\ likelihood) = -0.5 * \chi^2$.  This indicates a close connection between the "standard" way of looking at a data vs. model (minimize $\chi^2$) vs the Bayesian way of doing things (maximize posterior likelihood).  The main difference is in $p(\theta)$, or the prior distribution of your parameters.  This is the a bone of contention in Bayesian analysis, but it enables a whole new type of data modeling if used appropriately.  It also allows us to use our understanding about the world to impact our analysis, for good or for evil.

## Comparing Frequentist and Bayesian Approaches:

The frequentist way of looking at things is to think of probabilities in terms of repeated measurements.  You make a series of repeated measurements, and uncover the probability distribution through the number of times a given value comes up, or by the moments of all the numbers (mean, standard deviation).  Underlying this is the assumption that there is a single true value you are trying to uncover, and the noise comes from your system.  Another way of saying it: probabilities and frequencies are directly related.  A downside of this view is that science infrequently proceeds by making repeated measurements of the exact same effect in the exact same way.  

 * One way of reducing this is to say its hypothesis testing: compare to the null hypothesis; what is the p-value or significance of your results.
 
 * In terms of simulation, you can draw data from the model: $p(y|\theta)$.  
 
 * The studies are repeatable, the underlying parameters are fixed.  
  
 * In repeated sampling, you have 68% of the intervals covering the true parameter.

The Bayesian way of looking at things is to think of probabilities in terms of confidence limits.  Given your observation of the data, your confidence in the model is such and such.  In this case probabilities are related to our own knowledge of an event, and yield probaility distributions of parameter values.  Which to me seems more natural, but is a bit more difficult to represent in practice (not a single number with error bars, but a full posterior distribution).  

 * The reductionist way of thinking about this is that its interval estimation.  
 
 * In terms of simulation, you can draw models from the data: $p(\theta | y)$.  
 
 * The data are fixed.  
 
 * With probability 68%, the parameter is in the interval.  

## Practical Example: Fitting a Line

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(666)

def randomData(m, b, sig, npts = 100):
    x    = np.random.random(npts)
    mean = m * x + b
    y    = stats.norm(loc=mean, scale=sig).rvs(size=mean.shape)
    return x, y

mTrue = 3.2
bTrue = 1.7
sTrue = 0.3
xRand, yRand = randomData(mTrue, bTrue, sTrue)
dyRand = sTrue * np.ones_like(xRand)
plt.errorbar(xRand, yRand, yerr=dyRand, fmt="ro")
xModel = np.arange(0, 1.01, 0.1)
yModel = mTrue * xModel + bTrue
plt.plot(xModel, yModel, "k-")
plt.show()
print np.sum(((yRand - xRand*mTrue - bTrue)/dyRand)**2)
```

### Linear Fitting: Solve for Z below

How to solve:

The **normal equations** for unweighted least squares are:
  $$Y = A Z$$
  $$A^T Y = A^T A Z$$
  $$Z = [A^T A]^{-1} A^T Y$$

The **normal equations** for _weighted_ least squares use data covariance matrix $C$:
  $$Y = A Z$$
  $$C^{-1} Y = C^{-1} A Z $$
  $$A^T C^{-1} Y = A^T C^{-1} A Z $$
  $$Z = [A^T C^{-1} A]^{-1} [A^T C^{-1} Y]$$
  Parameter covariance matrix = $[A^T C^{-1} A]^{-1}$

Note that this minimizes:
  $$\chi^2 = [Y - AZ]^T C^{-1} [Y - AZ]$$

```python
A   = np.vstack((np.ones_like(xRand), xRand)).T
C   = np.diag(dyRand**2)
cov = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))
bLs, mLs = np.dot(cov, np.dot(A.T, np.linalg.solve(C, yRand)))
print "Linear fit"
print " b = %.3f +/- %.3f" % (bLs, np.sqrt(cov[0][0]))
print " m = %.3f +/- %.3f" % (mLs, np.sqrt(cov[1][1]))
fig = plt.figure()
sp = fig.add_subplot(111)
sp.errorbar(xRand, yRand, yerr=dyRand, fmt="ro")
sp.plot(xModel, mLs * xModel + bLs, "k-")
plt.show()
```

### Non-linear Fitting:

In this case, you define your $\chi^2$ function, and then use methods whose sole purpose is to minimize this function value by stepping through the model parameters that generate the $\chi^2$ (in the case of a line, slope and intercept).  This is necessary when the model you are fitting is not linear (e.g. sin, cos, exp, log, log10), and you can't proceed by simple matrix math.  

There are sophisticated methods that are used to find the uncertainties on parameters, but basically once the function finds its minimum, it steps around parameter space until it finds a $\Delta \chi^2$ value of 1.0, 4.0, 9.0 for the 1,2,3-$\sigma$ error bars, _when you are fitting 1 parameter_.  When you are fitting N parameters, the rule is that $\Delta \chi^2$ is itself distributed as a $\chi^2$ distribution with N degrees of freedom.  So to get your 68.3% confidence level when fitting 2 parameters, you want to find $\Delta \chi^2$ of 2.3, for 95.4% $\Delta \chi^2$ of 6.17, etc.  And note that these will in general be N-dimensional _contours_, but people typically just report the uncertainty on each parameter as if they were uncorrelated.  This is also typically a bad assumption.

To close the loop on model complexity, you expect that when _adding_ N terms to your model, $\Delta \chi^2$ follows the same $\chi^2$ distribution with N degrees of freedom.  So if you add 3 terms to your model, to be 3-sigma (99.73%) confident that all of those terms are needed, you want $\chi^2$ to improve by at least 14.2 (and typically much more).  See Numerical Recipes Section 15.6.

```python
import scipy.optimize as opt
def chi2(params, *args):
    b, m = params
    xf, yf, dyf = args
    prediction = m * xf + b
    return np.sum( ((prediction - yf) / dyf)**2 )

# You must always provide a starting guess for a minimization package.  If this is far from the true solution,
# the process may fail entirely
guess = [bTrue, mTrue]
result = opt.minimize(chi2, guess, args=(xRand, yRand, dyRand))
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result.x[0], np.sqrt(result.hess_inv[0][0]))
print " m = %.3f +/- %.3f" % (result.x[1], np.sqrt(result.hess_inv[1][1]))
print " chi2 = %.3f; chi2/dof = %.3f" % (result.fun, result.fun/(len(xRand)-len(result.x)))
```

The uncertainties are considerably different than the linear fit!  The Hessian inverse is not a good error bar estimator.  Try another method (note that opt.leastsq requires a slightly different return value than opt.minimize).

```python
def chi2(params, *args):
    b, m = params
    xf, yf, dyf = args
    prediction = m * xf + b
    return ((prediction - yf) / dyf)

guess = [bTrue, mTrue]
result = opt.leastsq(chi2, guess, args=(xRand, yRand, dyRand), full_output=1)
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result[0][0], np.sqrt(result[1][0][0]))
print " m = %.3f +/- %.3f" % (result[0][1], np.sqrt(result[1][1][1]))
chi2sum = np.sum(result[2]["fvec"]**2)
print " chi2 = %.3f; chi2/dof = %.3f" % (chi2sum, chi2sum/(len(xRand)-len(result[0])))
print " cov:"
print result[1]
```

```python
from astroML.resample import bootstrap

def resample(indices):
    return np.array([opt.leastsq(chi2, guess, args=(xRand[idx], yRand[idx], dyRand[idx]))[0] for idx in indices])

samples = bootstrap(xRand, 1000, resample, pass_indices=True)
print "Bootstrap"
print " b = %.3f +/- %.3f" % (np.mean(samples[:,0]), np.std(samples[:,0]))
print " m = %.3f +/- %.3f" % (np.mean(samples[:,1]), np.std(samples[:,1]))
```

What does the $\chi^2$ surface look like?

```python
npts = 20
chimat = np.empty((npts, npts))
bbest = np.mean(samples[:,0])
mbest = np.mean(samples[:,1])
brange = np.linspace(bbest-0.4, bbest+0.4, npts)
mrange = np.linspace(mbest-0.4, mbest+0.4, npts)
for i, bi in enumerate(brange):
    for j, mj in enumerate(mrange):
        chimat[i,j] = np.sum(chi2((bi, mj), xRand, yRand, dyRand)**2)
        
```

### What About When the Error Bars are Wrong?

```python
sWrong = 0.2
dyRand = sWrong * np.ones_like(xRand)
A   = np.vstack((np.ones_like(xRand), xRand)).T
C   = np.diag(dyRand**2)
cov = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))
bLs, mLs = np.dot(cov, np.dot(A.T, np.linalg.solve(C, yRand)))
print "Linear fit"
print " b = %.3f +/- %.3f" % (bLs, np.sqrt(cov[0][0]))
print " m = %.3f +/- %.3f" % (mLs, np.sqrt(cov[1][1]))
result = opt.leastsq(chi2, guess, args=(xRand, yRand, dyRand), full_output=1)
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result[0][0], np.sqrt(result[1][0][0]))
print " m = %.3f +/- %.3f" % (result[0][1], np.sqrt(result[1][1][1]))
chi2sum = np.sum(result[2]["fvec"]**2)
print " chi2 = %.3f; chi2/dof = %.3f" % (chi2sum, chi2sum/(len(xRand)-len(result[0])))
samples = bootstrap(xRand, 100, resample, pass_indices=True)
print "Bootstrap"
print " b = %.3f +/- %.3f" % (np.mean(samples[:,0]), np.std(samples[:,0]))
print " m = %.3f +/- %.3f" % (np.mean(samples[:,1]), np.std(samples[:,1]))
```

A couple things to note:

* In this case (all the data having the same uncertainties) the "best fits" are correct

* The uncertinaty estimates are off for both the linear and non-linear fits

* Bootstrap does a decent job of recovering the parameter uncertainties when the data error bars are wrong

* Data error bars are frequently wrong

* You get a sense that something is wrong with your model from the large chi2/dof

### Fitting for the Error Bars

Not possible to do linearly.  But you can do so easily by appropriately modifying the $\chi^2$ function.

```python
def chi2_sigma(params, *args):
    b, m, sig = params
    xf, yf, dyf = args
    sigsq = dyf**2 + sig**2
    prediction = m * xf + b
    return (prediction - yf)/ np.sqrt(sigsq)

guess = [bTrue, mTrue, 0.2]
result = opt.leastsq(chi2_sigma, guess, args=(xRand, yRand, dyRand), full_output=1)
print result[4]
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result[0][0], np.sqrt(result[1][0][0]))
print " m = %.3f +/- %.3f" % (result[0][1], np.sqrt(result[1][1][1]))
print " sig = %.3f +/- %.3f" % (result[0][2], np.sqrt(result[1][2][2]))
chi2sum = np.sum(result[2]["fvec"]**2)
print " chi2 = %.3f; chi2/dof = %.3f" % (chi2sum, chi2sum/(len(xRand)-len(result[0])))
```

Leastsq fails here!  http://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.leastsq.html#scipy.optimize.leastsq What about minimize?

```python
def chi2_sigma(params, *args):
    b, m, sig = params
    xf, yf, dyf = args
    sigsq = dyf**2 + sig**2
    prediction = m * xf + b
    return np.sum((prediction - yf)**2/ sigsq)

guess = [bTrue, mTrue, 0.2]
result = opt.minimize(chi2_sigma, guess, args=(xRand, yRand, dyRand))
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result.x[0], np.sqrt(result.hess_inv[0][0]))
print " m = %.3f +/- %.3f" % (result.x[1], np.sqrt(result.hess_inv[1][1]))
print " sig = %.3f +/- %.3f" % (result.x[2], np.sqrt(result.hess_inv[2][2]))
print " chi2 = %.3f; chi2/dof = %.3f" % (result.fun, result.fun/(len(xRand)-len(result.x)))

def resample_sigma(indices):
    return np.array([opt.minimize(chi2_sigma, guess, args=(xRand[idx], yRand[idx], dyRand[idx])).x for idx in indices])
samples = bootstrap(xRand, 100, resample_sigma, pass_indices=True)
print "Bootstrap"
print " b = %.3f +/- %.3f" % (np.mean(samples[:,0]), np.std(samples[:,0]))
print " m = %.3f +/- %.3f" % (np.mean(samples[:,1]), np.std(samples[:,1]))
print " sig = %.3f +/- %.3f" % (np.mean(samples[:,2]), np.std(samples[:,2]))
```

#### And now shit starts to off the rails...  Optimize technically works, but gives the wrong error bars (which indicates correct error bar estimation is what caused leastsq to fail).  And note that $\chi^2$ = 0.  Basically, the sigmas are being inflated so that the denominator of $\chi^2$ goes to infinity.  Note that bootstrapping does a decent job of recovering from this, but provides a poor estimate of sigma and its uncertainty.

The standard $\chi^2$ definition is insufficient once we get to this type of analysis.  Can we fix this by using the full likelihood?  This will allow $ln(2\pi\sigma_n^2)$ to ask as a "restoring force" in the analysis.

```python
sWrong = 0.2
dyRand = sWrong * np.ones_like(xRand)
def lnlike(params, *args):
    b, m, sig = params
    xf, yf, dyf = args
    prediction = m * xf + b
    sigsq = dyf**2 + sig**2
    # We could also consider someting where the variance is underestimated by a fractional amount:
    # sigsq = dyf**2 + (f * prediction)**2
    
    t1 = -0.5 * np.sum( (prediction-yf)**2/sigsq )
    t2 = -0.5 * np.sum( np.log(2 * np.pi * sigsq) )
    return t1 + t2
def chi2_like(params, *args):
    return -2 * lnlike(params, *args)

guess = [bTrue, mTrue, 0.2]
result = opt.minimize(chi2_like, guess, args=(xRand, yRand, dyRand))
print "Nonlinear fit"
print " b = %.3f +/- %.3f" % (result.x[0], np.sqrt(result.hess_inv[0][0]))
print " m = %.3f +/- %.3f" % (result.x[1], np.sqrt(result.hess_inv[1][1]))
print " sig = %.3f +/- %.3f" % (result.x[2], np.sqrt(result.hess_inv[2][2]))
print " chi2 = %.3f; chi2/dof = %.3f" % (result.fun, result.fun/(len(xRand)-len(result.x)))

def resample_like(indices):
    return np.array([opt.minimize(chi2_like, guess, args=(xRand[idx], yRand[idx], dyRand[idx])).x for idx in indices])
samples = bootstrap(xRand, 100, resample_like, pass_indices=True)
print "Bootstrap"
print " b = %.3f +/- %.3f" % (np.mean(samples[:,0]), np.std(samples[:,0]))
print " m = %.3f +/- %.3f" % (np.mean(samples[:,1]), np.std(samples[:,1]))
print " sig = %.3f +/- %.3f" % (np.mean(samples[:,2]), np.std(samples[:,2]))
```

Note that sigma is somewhat overestimated; we expect its value to be $\sqrt(0.3^2-0.2^2) = 0.224$.  We can see that the variance is being somewhat overestimated by $\chi^2/dof < 1$.

## How Does Bayes Help?

We use simulation methods to approximate the target posterior distribution $p(\theta | y)$.  Markov Chain Monte Carlo (MCMC) methods are used to sample from the posterior in a way
that the distribution of chain steps are close to the stationary distribution of $p(\theta | y)$.
This is a sampling of model parameters, conditional on a previous sample of model parameters and
the relative values of $p(\theta | y)$ at the 2 steps.

There are 2 primary methods for MCMC: Gibbs sampling and Metropolis-Hastings algorithms. MH algorithms include a random walk 
through parameter space with an acceptance rule tuned to converge to $p(\theta | y)$.  Gibbs samplers
are a subset of MH samplers with a 100% acceptance rate, but only work for a subset of
(conditionally conjugate?) models.

Using MCMC methods is pretty straightforward these days.  We use the emcee package, and start a set of "walkers" through parameter space, initialized in some ball around our guess.  After a "burn-in" phase, they are intended to sample the space with a density that is proportional to the posterior likelihood.  This is somewhat analagous to optimization, except that _the optimal point is not guaranteed to be sampled_.  Instead, the joint distribution of parameter values is returned; a marginalized distribution can be obtained by looking at the distribution of only 1 parameter.  You can use the step with the highest likelihood as a point estimate of $p(\theta | y)$.

You always want to do tests for convergence: mixing and stationarity, within a chain and between chains (e.g. the Rubin-Gelman R^hat statistic, which estimates how much the knowledge of your parameters would improve if nsteps -> $\infty$).  Also important are the effective number of samples, which can be determined by looking at the autocorrelation length of the chains.  The effective number of samples is approximately the total number of steps divided by the autocorrelation length (See also BDA 11.5 for a more comprehensive definition).

```python
import acor
# Autocorrelation for a given parameter, for each walker
for n in range(nwalkers):
    tau, mean, sigma = acor.acor(sampler.chain[n,:,0])
    print n, tau, nstep/tau
# Autocorrelation for all parameters including appending the chains
tau, mean, sigma = acor.acor(sampler.flatchain[:,0])
print "ALL", tau, nstep/tau
```

The "marginal posterior distribution" for each parameter $\theta_i$ is defined as $p(\theta_i | y)$.  That is, it is looking at only 1 dimension (or slice) of the joint distribution $p(\theta | y)$.  When people talk about "marginalizing" over nuisance parameters, mathematically it means (for a 2-dimensional distribution) : $p(\theta_1|y) = \int p(\theta_1, \theta_2 | y) d\theta_2$.  In terms of Markov chain simulation, it practically means collapsing all samples of $\theta_1$ into a single histogram, ignoring correlations with $\theta_2$.

### How to set priors:

```python
def lnprior(params):
    b, m, sig = params
    if sig < 0.3 or sig > 0.4:
        return -np.inf
    return 0.0

def lnlike_prior(params, *args):
    b, m, sig = params
    lnp = lnprior(params)
    
    xf, yf, dyf = args
    prediction = m * xf + b
    sigsq = dyf**2 + sig**2
    t1 = -0.5 * np.sum( (prediction-yf)**2/sigsq )
    t2 = -0.5 * np.sum( np.log(2 * np.pi * sigsq) )
    return t1 + t2 + lnp

guess  = [1.6, 3.2, 0.35]
eguess = [np.array(guess) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlike_prior, args=(xRand, yRand, dyRand))
eguess, prob, state = sampler.run_mcmc(eguess, nburn)
sampler.reset()
efinal, prob, state = sampler.run_mcmc(eguess, nstep, rstate0=state)
probs    = sampler.lnprobability.reshape(np.product(sampler.lnprobability.shape))
sortIdx  = np.argsort(probs)[::-1]
bestPars = sampler.flatchain[sortIdx][0]
print "Maximum a-posteriori estimate:", bestPars
print " b = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,0], 50),
                              .741*(np.percentile(sampler.flatchain[:,0], 75) - np.percentile(sampler.flatchain[:,0], 25))) 
print " m = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,1], 50),
                              .741*(np.percentile(sampler.flatchain[:,1], 75) - np.percentile(sampler.flatchain[:,1], 25))) 
print " sig = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,2], 50),
                                .741*(np.percentile(sampler.flatchain[:,2], 75) - np.percentile(sampler.flatchain[:,2], 25))) 
```

### The important thing is to be able to define your entire posterior probability distribution, via a generative model and its likelihood function and a prior.  You then sample from this full posterior with your MCMC sampler.  Turning a standard $\chi^2$ function into a likelihood function is an acceptable _but naive_ way to transition from a classical frequentist to Bayesian analysis (guilty).  The full buy-in comes when you redefine your likelihood function to better reflect the underlying physical processes that generated your data.  This is at the leading edge of a new (but growing) trend in astrophysics of reimagining the process of data analysis using Bayesian techniques.

## Hierarchical Modeling:

We use hierarchical, or _multi-level_ modeling, to build up more complicated models than are easily allowed for in 
frequentist analysis.  An important aspect of this is including unknown terms, e.g. variance, and marginalizing over
them (i.e. collapsing the histogram of variance vs. parameter down on the x-axis).  These nuisance parameters may
be modeled so that their effects and interactions with the data are captured as part of the model, even if they are 
not directly interesting to the analysis.

A second advantage is the use of "partial" pooling of the data.  This presents a third option for jointly fitting multiple realizations of the same effect.  The first is that you may fit each dataset separately.  This is be a low bias, high variance method, especially when some of the datasets have small numbers of points.  The second is that you may fit each dataset jointly.  This is a high bias, low variance method as each dataset is constrained to have the exact same response.  This third hierarchical method is a technique to moderate the bias<->variance tradeoff by proposing that each dataset comes from a common population; the modeling of that population then becomes the key focus of the fitting.  **Its important to distinguish that what you are really after are population means and standard deviations, not sample means and standard deviations.**

For an ensemble of linear relationships that are drawn from the same population, but might have different slopes, a multi-level model is designed as follows:

#### Measurement Level Model:

Model of how the data are generated from the model.  E.g. y = $N(m x + b; \sigma^2)$

#### Population Level (or Prior) Model

Model of the underlying population.  E.g. $m = N(\mu, \tau^2)$.  Here $\tau$ is a hyperparameter of your model and
represents the underlying stochastic variation of the process that generates the slopes for your samples.   I think
the priors for $b$ and $\sigma^2$ technically go here as well, but thats not particularly important.

#### Hyper-prior Level

Sets priors for the population level parameters of your model, $\mu$ and $\tau$.  Flat priors are typically used to avoid contention.  People also typically use 1-sided t-distributions (e.g. half-Cauchy) on sigmas (or inverse gamme on variances).

#### Lets set one up in code, including random numbers of points per sample:

```python
def randomData(mu, tau, b, sig, nptsMax = 20):
    npts = np.random.randint(3, nptsMax)
    x    = np.random.random(npts)
    m    = stats.norm(loc=mu, scale=tau).rvs()
    mean = m * x + b
    y    = stats.norm(loc=mean, scale=sig).rvs(size=mean.shape)
    return x, y, m
ncurves = 12
muTrue  = 3.2
tauTrue = 0.7 # This is the width of the distribution, i.e. sigma
bTrue   = 1.7
allData = []
fig = plt.figure(figsize=(19, 5))
for i in range(ncurves):
    xRand, yRand, mRand = randomData(muTrue, tauTrue, bTrue, sTrue)
    dyRand = sTrue * np.ones_like(xRand)
    sp = fig.add_subplot(1, ncurves, i+1)
    sp.set_title("%.3f" % (mRand))
    sp.errorbar(xRand, yRand, yerr=dyRand, fmt="ro")    
    allData.append( (xRand, yRand, dyRand) )
```

### Unpooled Analysis: Each model fit independently

```python
def lnprior(params):
    b, m, lnsig = params
    if lnsig < -10.0 or lnsig > 0.0:
        return -np.inf
    return 0

def lnlike(params, *args):
    b, m, lnsig = params
    xf, yf, dyf = args
    
    lnp = lnprior(params)
    
    prediction = m * xf + b
    sigsq = dyf**2 + np.exp(lnsig)**2
    t1 = -0.5 * np.sum( (prediction-yf)**2/sigsq )
    t2 = -0.5 * np.sum( np.log(2 * np.pi * sigsq) )
    return t1 + t2 + lnp

guess = (bTrue, mRand, np.log(1e-1))
fig = plt.figure(figsize=(19, 5))
bs = []
ms = []
lnsigs = []
for n in range(ncurves):
    ndim, nwalkers, nburn, nstep = len(guess), 2*len(guess), 1000, 10000
    eguess = [np.array(guess) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
    sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlike, args=(allData[n][0], allData[n][1], allData[n][2]))
    eguess, prob, state = sampler.run_mcmc(eguess, nburn)
    sampler.reset()
    efinal, prob, state = sampler.run_mcmc(eguess, nstep, rstate0=state)
    print "Curve", n
    probs    = sampler.lnprobability.reshape(np.product(sampler.lnprobability.shape))
    sortIdx  = np.argsort(probs)[::-1]
    bestPars = sampler.flatchain[sortIdx][0]
    bs.append(bestPars[0])
    ms.append(bestPars[1])
    lnsigs.append(bestPars[2])
    print "Maximum a-posteriori estimate:", bestPars
    print " b = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,0], 50),
                                  .741*(np.percentile(sampler.flatchain[:,0], 75) - np.percentile(sampler.flatchain[:,0], 25))) 
    print " m = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,1], 50),
                                  .741*(np.percentile(sampler.flatchain[:,1], 75) - np.percentile(sampler.flatchain[:,1], 25))) 
    print " sig = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,2], 50),
                                    .741*(np.percentile(sampler.flatchain[:,2], 75) - np.percentile(sampler.flatchain[:,2], 25))) 
    
    sp = fig.add_subplot(1, ncurves, n+1)
    sp.errorbar(allData[n][0], allData[n][1], yerr=allData[n][2], fmt="ro")    
    for b, m, sig in sampler.flatchain[np.random.randint(len(sampler.flatchain), size=100)]:
        sp.plot(xModel, m * xModel + b, "k-", alpha=0.05)
```

Recall that we expect the mean of M to be 3.2 with an RMS of 0.7.  The mean is close but the RMS is an overestimate.  It is very difficult in these cases to estimate your uncertainty on the width of the distribution.  I.e. how many sigma away from 0.7 is 0.85?

### Pooled Analysis: Common slope

```python
def lnprior(params):
    b, m, lnsig = params
    if lnsig < -10.0 or lnsig > 0.0:
        return -np.inf
    return 0

def lnlike(params, *args):
    b, m, lnsig = params
    allData = args
    
    lnp = lnprior(params)
    
    lnl = 0
    for i in range(ncurves):
        xf, yf, dyf = allData[i]
        prediction = m * xf + b
        sigsq = dyf**2 + np.exp(lnsig)**2
        t1 = -0.5 * np.sum( (prediction-yf)**2/sigsq )
        t2 = -0.5 * np.sum( np.log(2 * np.pi * sigsq) )
        lnl += t1 + t2
    return lnl + lnp

guess = (bTrue, mRand, np.log(1e-1))
ndim, nwalkers, nburn, nstep = len(guess), 2*len(guess), 1000, 10000
eguess = [np.array(guess) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlike, args=(allData))
eguess, prob, state = sampler.run_mcmc(eguess, nburn)
sampler.reset()
efinal, prob, state = sampler.run_mcmc(eguess, nstep, rstate0=state)
probs    = sampler.lnprobability.reshape(np.product(sampler.lnprobability.shape))
sortIdx  = np.argsort(probs)[::-1]
bestPars = sampler.flatchain[sortIdx][0]
print "Maximum a-posteriori estimate:", bestPars
print " b = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,0], 50),
                              .741*(np.percentile(sampler.flatchain[:,0], 75) - np.percentile(sampler.flatchain[:,0], 25))) 
print " m = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,1], 50),
                              .741*(np.percentile(sampler.flatchain[:,1], 75) - np.percentile(sampler.flatchain[:,1], 25))) 
print " ln sig = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,2], 50),
                                   .741*(np.percentile(sampler.flatchain[:,2], 75) - np.percentile(sampler.flatchain[:,2], 25))) 
fig = plt.figure(figsize=(15, 5))
for b, m, sig in sampler.flatchain[np.random.randint(len(sampler.flatchain), size=100)]:
    for n in range(ncurves):
        sp = fig.add_subplot(1, ncurves, n+1)
        sp.errorbar(allData[n][0], allData[n][1], yerr=allData[n][2], fmt="ro")    
        sp.plot(xModel, m * xModel + b, "k-", alpha=0.05)
```

Here the RMS on the slope (0.17) is very small compared to the designed width 0.7.  Again, we don't have an easy way to infer any uncertainties on the width of the distribution.  In fact its tricky to phrase what we want in this representation: it would be the uncertainty on the RMS uncertainty on the slope.  When what we are truly after is the width of the distribution of slopes.

### Partial Pooling: Hierarchical Analysis

```python
def lnprior(params):
    ms    = params[:ncurves]
    mu    = params[ncurves]
    lntau = params[ncurves+1]
    b     = params[ncurves+2]
    lnsig = params[ncurves+3]

    if lntau < -10.0 or lntau > 0.0:
        return -np.inf
    
    if lnsig < -10.0 or lnsig > 0.0:
        return -np.inf
    
    return 0

def lnlike(params, *args):
    ms    = params[:ncurves]
    mu    = params[ncurves]
    lntau = params[ncurves+1]
    b     = params[ncurves+2]
    lnsig = params[ncurves+3]
    allData = args
    
    # Hyper prior level
    lnp = lnprior(params)
    
    # Population level prior
    lnlp = np.sum([stats.norm.logpdf(m, mu, np.exp(lntau)) for m in ms])
    
    # Measurement Level
    lnl = 0
    for i in range(ncurves):
        xf, yf, dyf = allData[i]
        prediction = ms[i] * xf + b
        sigsq = dyf**2 + np.exp(lnsig)**2
        t1 = -0.5 * np.sum( (prediction-yf)**2/sigsq )
        t2 = -0.5 * np.sum( np.log(2 * np.pi * sigsq) )
        lnl += t1 + t2
    return lnl + lnlp + lnp


# We include a slope per data set
guess  = ncurves * [muTrue,]
# With a common parent population to link them, mean and standard deviation
guess += [muTrue, np.log(tauTrue)]
# A joint intercept
guess += [bTrue,]
# And an additional noise term, for completeness
guess += [np.log(1e-1,)]

ndim, nwalkers, nburn, nstep = len(guess), 2*len(guess), 1000, 10000
eguess = [np.array(guess) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlike, args=(allData))
eguess, prob, state = sampler.run_mcmc(eguess, nburn)
sampler.reset()
efinal, prob, state = sampler.run_mcmc(eguess, nstep, rstate0=state)
probs    = sampler.lnprobability.reshape(np.product(sampler.lnprobability.shape))
sortIdx  = np.argsort(probs)[::-1]
bestPars = sampler.flatchain[sortIdx][0]
print "Maximum a-posteriori estimate:", bestPars
for n in range(ncurves):
    print " m%d = %.3f +/- %.3f" % (n, np.percentile(sampler.flatchain[:,n], 50),
                                      .741*(np.percentile(sampler.flatchain[:,n], 75) - np.percentile(sampler.flatchain[:,n], 25))) 
print
print " mu = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,ncurves], 50),
                              .741*(np.percentile(sampler.flatchain[:,ncurves], 75) - np.percentile(sampler.flatchain[:,ncurves], 25))) 
print " tau = %.3f +/- %.3f" % (np.exp(np.percentile(sampler.flatchain[:,ncurves+1], 50)),
                                0.741*(np.exp(np.percentile(sampler.flatchain[:,ncurves+1], 75)) - np.exp(np.percentile(sampler.flatchain[:,ncurves+1], 25))))
print
print " b = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,ncurves+2], 50),
                              .741*(np.percentile(sampler.flatchain[:,ncurves+2], 75) - np.percentile(sampler.flatchain[:,ncurves+2], 25))) 
print " ln sig = %.3f +/- %.3f" % (np.percentile(sampler.flatchain[:,ncurves+3], 50),
                                  .741*(np.percentile(sampler.flatchain[:,ncurves+3], 75) - np.percentile(sampler.flatchain[:,ncurves+3], 25))) 
```

Note that we are fitting directly for the terms that are imporant to us: the mean of the slope distribution (3.12 vs. the input 3.2) and the width of this distribution (0.78 vs. the input 0.7).  Importantly, we have confidence regions for both parameters.  Of note the intercept is within 1-sigma of the inputs (1.7) and the additional noise term is negligible ($10^{-6}$ compared to _no_ additional noise).
